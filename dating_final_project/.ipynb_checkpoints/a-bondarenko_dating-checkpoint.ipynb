# Проект «Дейтинговое приложение»
Есть крупное дейтинговое приложение.

Помимо базовых функций, в нём также имеется премиум-подписка, которая дает доступ к ряду важных дополнительных возможностей. Был проведен A/B тест, в рамках которого для новых пользователей из нескольких стран была изменена стоимость премиум-подписки* при покупке через две новые платежные системы. При этом стоимость пробного периода оставалась прежней.

Нам нужно проверить был ли эксперимент успешен в целом.

*Деньги за подписку списываются ежемесячно до тех пор, пока пользователь её не отменит.

## Описание данных
Всего есть три группы: 
- `test` — тестовая 
- `control_1` — контрольная 1 
- `control_2` — контрольная 2

Для каждой из них есть 2 вида файлов (всего 6):

**`users_*.csv`** — информация о пользователях:

- `uid` — идентификатор пользователя
- `age` — возраст
- `attraction_coeff` — коэффициент привлекательности (от 0 до 1000, $\frac{лайки}{просмотры}*1000$)
- `coins` — число монеток (внутренняя валюта)
- `country` — страна
- `visit_days` — в какие дни после регистрации пользователь посещал приложение (например, в первый, а затем в седьмой)
- `gender` — пол
- `age_filter_start` — фильтр поиска, минимальное значение
- `age_filter_end` — фильтр поиска, максимальное значение
- `views_count` — число просмотров
- `was_premium` — был ли когда-либо премиум (либо пробный период премиум-статуса, либо купленный за деньги)
- `is_premium` — является ли премиум
- `total_revenue` — нормированная выручка

**`transactions_*.csv`** — информация о платежах пользователей:

- `uid` — идентификатор пользователя
- `country` — страна
- `joined_at` — дата и время регистрации
- `paid_at` — дата и время покупки
- `revenue` — нормированная выручка
- `payment_id` — идентификатор платежа
- `from_page` — откуда пользователь перешел на страницу оплаты
- `product_type` — тип продукта:
    - `trial_premium` — пробная премиум-подписка,
    - `premium_no_trial` — премиум-подписка без пробной, 
    - `coins` — подписка за внутреннюю валюту, 
    - `other_type` — другое

Записей после 15 ноября нет, а это значит, что диапазон данных охватывает период в месяц, что уже похоже на правду. Теперь необходимо сделать 2 вещи:
1. Посмотреть, нет ли такой же ситуации в двух других таблицах
2. Исправить проблему, благо мы можем сделать это безболезненно из-за удобных дат начала и окончания эксперимента

```python
trans_control_1[trans_control_1['paid_at'].dt.month == 11].sort_values(by='paid_at')
```

```python
trans_control_2[trans_control_2['paid_at'].dt.month == 11].sort_values(by='paid_at')
```

Всё то же самое, поэтому устранять ошибки мы будем сразу для всего семейства данных `transactions_*.csv`. Везде поменяем местами дату и месяц, **для каждого 11 числа**. Нам очень повезло, потому что единственное настоящее 11 число — это 11.11, поэтому наши изменения никак не затронут данные в октябре

```python
for i in [trans_test, trans_control_1, trans_control_2]:
    i.loc[i['joined_at'].dt.day == 11, 'joined_at'] = i.loc[i['joined_at'].dt.day == 11, 'joined_at'] \
        .apply(lambda x: x.replace(day=x.month, month=11))
    i.loc[i['paid_at'].dt.day == 11, 'paid_at'] = i.loc[i['paid_at'].dt.day == 11, 'paid_at'] \
        .apply(lambda x: x.replace(day=x.month, month=11))
```

## Выбор ключевых метрик 
Поскольку изменения затронули только стоимость *премиума без триала*, то логично будет измерять показатели, которые в первую очередь связаны конкретно с ним. Я выбрал 3 метрики — 1 главную и 2 прокси:

**Главная**
1. Средний доход на уникального пользователя премиума без триала. Это аналог ARPPU, только направленный на конкретную механику монетизации. Логика простая: изменили цену — смотрим на деньги, которые принесли юзеры. Наша задача, чтобы эта метрика статистически значимо увеличилась. Назомём её Average Revenue Per Premium User — **ARPPrU**

**Прокси или guardrail**
1. Конверсия из пользователя в покупателя премиума без триала. Если мы как-то меняем цену в ту или иную сторону, то у нас может измениться количество юзеров, которые готовы что-то по этой цене покупать. Если уроним эту метрику, но стат. значимо подымем средний доход  на пользователя премиума, то не страшно. Однако, если ARPPrU не увеличится, а конверсия упадёт, то это повод не раскатывать новую ценовую политику на всех. Назовём её Conversion Rate To Premium — **CRPr**
2. Средний доход на любого уникального пользователя, то есть самый обычный **ARPU**. Логика в том, чтобы посмотреть, а не каннибализировало ли изменение цены остальные продукты в приложении. То есть средний доход в конкретной механике мы подняли, а остальные из-за этого просели и в итоге мы в среднем стали недополучать денег с одного юзера. Наша задача не уронить эту метрику при любых раскладах

**От чего отказался**
- Retention. У нас ежемесячная подписка, а сам эксперимент шёл всего месяц. Это слишком маленький срок для того, чтобы накопить эффект в этой метрике — я бы крутил такой эксперимент несколько месяцев, если, конечно, позволяли бы ресурсы бизнеса
- Доля отмен подписки среди людей, которые купили премиум во время эксперимента. Она не годится в качестве основного критерия успешности эксперимента, а прокси метрик у нас и так достаточно, причём более наглядных

### `revenue` vs `total_revenue`
Cравним сумму `revenue` c `total revenue` в разрезе каждого уникального юзера. Для этого напишем функцию, чтобы применить на все датафреймы

```python
def revenue_vs_total_revenue(merged_df):
    rev_vs_total_rev = merged_df.groupby('uid').agg({'total_revenue': 'first', 'revenue': 'sum'}).reset_index()
    sum_errors = rev_vs_total_rev.query("total_revenue != revenue")['uid']
    sum_errors_uids = merged_df.query("uid in @sum_errors")
    return sum_errors_uids
```

## Проверка системы сплитования
Чтобы убедиться в том, что все группы у нас разделены действительно случайно, я сравню распределения второстепенных данных, которые не влияют на наши целевые метрики. У нас три группы, для каждой нужно посмотреть распределения и проверить их статистическими тестами:
- для количественных параметров — связку из теста Левена и дисперсионного анализа (ANOVA). Нам достаточно будет убедиться, что средние всех групп равны, поэтому мы не будем углубляться в попарные тесты
- для категориальных параметров — тест хи-квадрат на независимость. Здесь хотим подтвердить, что распределение не зависит от группы тестирования

Количественные:
- `age`
- `attraction_coeff`
- `views_count`
- `coins`
- `count_visit_days`
- `age_filter_range`

Категориальные:
- `country`
- `gender`

Если во всех тестах мы подтведим нулевую гипотезу, то можно будет утверждать, что система сплитования отработала правильно

### Количественнные параметры
Гипотезы для теста Левена:
- $H_0$: дисперсии во всех группах равны
- $H_1$: дисперсии в каких-то группах не равны

Гипотезы для ANOVA:
- $H_0$: средние всех групп равны
- $H_1$: среднее хотя бы для одной группы отличается

Напишу общую функцию для вывода графиков и итогов статистических тестов

```python
def quantity_stat(parameter):
    quantity_par = task_all_data.groupby(['group', 'uid']).agg({parameter: 'first'}).reset_index()
    levene = pg.homoscedasticity(data=quantity_par, dv=parameter, group='group')
    var_test = pg.anova(data=quantity_par, dv=parameter, between='group')
    
    # Построение гистограмм для каждой группы
    plt.figure(figsize=(15, 5))
    
    for i, (group_name, group_data) in enumerate(quantity_par.groupby('group')):
        plt.subplot(1, 3, i+1)
        sns.histplot(group_data[parameter], kde=True, stat="density", common_norm=False, color=f'C{i}')
        plt.title(f'{group_name}')
        plt.xlabel(parameter)
        plt.ylabel('Density')
    
    plt.suptitle(f'Распределения {parameter}', y=1.02)
    plt.tight_layout()
    plt.show()
    
    # Вывод статистических тестов
    print("\nТест Левена:")
    display(levene)
    
    print("\nANOVA:")
    display(var_test)
```

Теперь рассмотрим наши параметры по очереди

```python
quantity_stat('age')
```

```python
quantity_stat('attraction_coeff')
```

```python
quantity_stat('views_count')
```

```python
quantity_stat('coins')
```

```python
quantity_stat('count_visit_days')
```

```python
quantity_stat('age_filter_range')
```

Мы не смогли отвернуть нулевые гипотезы ни для одного параметра

### Категориальные параметры
Гипотезы для теста хи-квадрат на независимость:
- $H_0$: распределение не зависит от группы
- $H_1$: распределение зависит от группы, то есть не случайно

Так же здесь нам нужно подтвердить нулевую гипотезу во всех случаях

Напишу функцию, а потом применю её к параметрам

```python
def quality_stat(parameter):
    quality_par = task_all_data.groupby(['group', 'uid']).agg({parameter: 'first'}).reset_index()
    cross = pd.crosstab(quality_par[parameter], quality_par['group'])
    exp, obs, stats = pg.chi2_independence(data=quality_par, x="group", y=parameter)
    
    print("Таблица сопряжённости:")
    display(cross)
    
    print("Тест хи-квадрат:")
    display(stats)
```

```python
quality_stat('country')
```

```python
quality_stat('gender')
```

Для качественных парметров тоже потвердили нулевые гипотезы

**Вывод**

Система сплитования отбработала правильно, можем приступать к проверке целевых метрик

## A/A тест
Чтобы тест можно было считать успешным, мы не должны увидеть статистически значимых различий в целевых показателях — их мы сформулировали в разделе «Выбор ключевых метрик»

### ARPPrU
Сразу напишу функцию, чтобы иметь возможность быстро вызвать её уже при А/B тестировании. Поскольку мы будем смотреть на средние значения, то будет использовать Т-тест

Гипотезы Т-теста:
- $H_0$: средние значения двух групп равны
- $H_1$: средние значения двух групп отличаются

```python
def arppru(group_1, group_2):
    arppru_df = task_all_data \
        .query(f"group.isin(['{group_1}', '{group_2}']) & bought_prem_no_trial == 'Yes'") \
        .groupby(['uid', 'group'], as_index=False) \
        .agg({'revenue': 'sum'})
    
    arppru_metrics = arppru_df.groupby('group') \
        .agg({'revenue': 'mean'}) \
        .rename(columns={'revenue': 'arppru'})
    
    arppru_levene = pg.homoscedasticity(data=arppru_df, dv='revenue', group='group')
    
    x = arppru_df.query(f"group == '{group_1}'")['revenue']
    y = arppru_df.query(f"group == '{group_2}'")['revenue']
    
    arppru_ttest = pg.ttest(x, y)
    
    display(arppru_df.head())
    
    # Построение гистограмм для каждой группы
    print("\nРаспределения:")
    plt.figure(figsize=(12, 4)) 

    for i, (group_name, group_data) in enumerate(arppru_df.groupby('group')):
        plt.subplot(1, 3, i+1)
        sns.histplot(group_data['revenue'], kde=True, stat="density", common_norm=False, color=f'C{i}')
        plt.title(f'{group_name}')
        plt.xlabel('Revenue')
        plt.ylabel('Density')

    plt.tight_layout()
    plt.show()

    print("\nЗначения метрики по группам:")
    display(arppru_metrics)

    # Вывод статистических тестов
    print("\nТест Левена:")
    display(arppru_levene)

    print("\nT-тест:")
    display(arppru_ttest)
```

```python
arppru('control_1', 'control_2')
```

**Вывод**

Средние в контрольных группах статистически значимо  не отличаются

### CRPr
Здесь конверсия, поэтому для статистического теста я использую хи-квадрат

```python
def crpr(group_1, group_2):
    crpr_df = task_all_data \
        .query(f"group.isin(['{group_1}', '{group_2}'])") \
        .groupby(['uid', 'group'], as_index=False) \
        .agg({'bought_prem_no_trial': 'first'})
    
    crpr_cross = pd.crosstab(crpr_df['bought_prem_no_trial'], crpr_df['group'])
    
    crpr_metrics = (crpr_cross.loc['Yes'] / (crpr_cross.loc['Yes'] + crpr_cross.loc['No'])) * 100
    crpr_metrics = crpr_metrics.to_frame().rename(columns={0: 'crpr'}).round(3)
    
    exp, obs, crpr_stats = pg.chi2_independence(data=crpr_df, x="group", y='bought_prem_no_trial')
    
    display(crpr_df.head())
    
    print("\nТаблица сопряжённости:")
    display(crpr_cross)
    
    print("\nЗначения метрики по группам:")
    display(crpr_metrics)
    
    print("\nТест хи-квадрат:")
    display(crpr_stats)
```

```python
crpr('control_1', 'control_2')
```

**Вывод**

Распределения в обоих группах не отличаются

### ARPU
В плане тестов всё то же самое, что и для ARPPrU

```python
def arpu(group_1, group_2):
    arpu_df = task_all_data \
        .query(f"group.isin(['{group_1}', '{group_2}'])") \
        .groupby(['uid', 'group'], as_index=False) \
        .agg({'total_revenue': 'first'})
    
    arpu_metrics = arpu_df.groupby('group') \
        .agg({'total_revenue': 'mean'}) \
        .rename(columns={'total_revenue': 'arpu'})
    
    arpu_levene = pg.homoscedasticity(data=arpu_df, dv='total_revenue', group='group')
    
    x = arpu_df.query(f"group == '{group_1}'")['total_revenue']
    y = arpu_df.query(f"group == '{group_2}'")['total_revenue']
    
    arpu_ttest = pg.ttest(x, y)
    
    display(arpu_df.head())
    
    # Построение гистограмм для каждой группы
    print("\nРаспределения:")
    plt.figure(figsize=(12, 4)) 

    for i, (group_name, group_data) in enumerate(arpu_df.groupby('group')):
        plt.subplot(1, 3, i+1)
        sns.histplot(group_data['total_revenue'], kde=True, stat="density", common_norm=False, color=f'C{i}')
        plt.title(f'{group_name}')
        plt.xlabel('Total Revenue')
        plt.ylabel('Density')

    plt.tight_layout()
    plt.show()
    
    print("\nЗначения метрики по группам:")
    display(arpu_metrics)
    
    print("\nТест Левена:")
    display(arpu_levene)
    
    print("\nТ-тест:")
    display(arpu_ttest)
```

```python
arpu('control_1', 'control_2')
```

**Вывод**

Средние в контрольных группах статистически значимо не отличаются

#### Итога A/A теста
Мы не нашли статистически значимых различий по ключевым метрикам, что говорит о правильной методологии нашего эксперимента
